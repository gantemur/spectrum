```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import collections as matcoll
import matplotlib.font_manager as fontman
```

```python
path = '/Library/Fonts/Arial Unicode.ttf'
prop = fontman.FontProperties(fname=path)
mpl.rcParams['font.family'] = prop.get_name()
```

```python
sigma=5.67e-8
S=1361
dplan=[.39,.723,1,1.524,5.203,9.539,19.18,30.06]
tplan=[452,726,285,230,120,88,59,48]
aplan=[.06,.75,.3,.29,.52,.47,.51,.41]
```

```python
def tem(r,a=0,e=1,f=.25):
    return ((1-a)*f*S/(r*r*e*sigma))**.25
```

```python
xres=np.linspace(.3,31,1000)
plt.figure(figsize=(15,10))
#for d in dplan: plt.axvline(x=d,linestyle='--')
plt.plot(xres,[tem(x) for x in xres],label='A=0, E=1')
plt.plot(xres,[tem(x,.3,.6) for x in xres],label='A=0.3, E=0.6')
plt.plot([d for d in dplan],[tem(d) for d in dplan],'bo')
#plt.plot([d for d in dplan],[tem(d,.5,.1) for d in dplan],'ro')
#plt.yscale('log')
plt.xlabel(u'Нар хүртэлх зай, одон орны нэгжээр',fontproperties=prop, size=14)
plt.ylabel(u'Температур, Кельвинээр',fontproperties=prop, size=14)
#plt.xticks(dplan)
plt.legend()
#plt.savefig("temp.png", dpi=150)
plt.show()
```

```python
xres=np.linspace(.3,31,1000)
xres1=np.linspace(.7,31,1000)
plt.figure(figsize=(15,10))
#plt.stem(dplan,tplan)
plt.plot(xres,[tem(x) for x in xres],label='A=0, E=1')
plt.plot(xres1,[tem(x,.75,.01) for x in xres1],label='A=0.75, E=0.01')
#plt.plot([d for d in dplan],[tem(d) for d in dplan],'b-')
#for i in range(len(dplan)): plt.vlines(dplan[i],tplan[i],tem(dplan[i]))
plt.plot([d for d in dplan],[t for t in tplan],'rx')
#plt.yscale('log')
plt.xlabel(u'Нар хүртэлх зай, одон орны нэгжээр',fontproperties=prop, size=14)
plt.ylabel(u'Температур, Кельвинээр',fontproperties=prop, size=14)
plt.legend()
#plt.savefig("tempd.png", dpi=150,bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.stem(dplan,[tplan[i]-tem(dplan[i]) for i in range(len(dplan))],markerfmt='.')
plt.xlabel(u'Нар хүртэлх зай, одон орны нэгжээр',fontproperties=prop, size=14)
plt.ylabel(u'Температур, Кельвинээр',fontproperties=prop, size=14)
#plt.savefig("temp.png", dpi=150)
#plt.yscale('log')
plt.show()
```
